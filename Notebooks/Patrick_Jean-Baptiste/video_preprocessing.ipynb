# Description: Video Preprocessing 

This notebook performs the initial video preprocessing which involves reading a video file, extracting video frames, and converting the video frames to grayscale.

```python
from IPython.display import Video
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Retrieve the name of the video file.

```python
vid_file = 'Video_Speech_Actor_01/Actor_01/01-01-03-01-01-01-01.mp4'
```

Play the video file.

```python
Video(vid_file)
```

Capture the video from the video file.

```python
cap = cv2.VideoCapture(vid_file)
```

Extract the video frames and convert them to grayscale.

```python
Vid_Frames = []

while cap.isOpened():
    # Read a video frame.
    ret, frame = cap.read()
    
    # If there is a video frame available for reading.
    if not ret:
        break

    # Convert the video frame from RGB to grayscale.
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Append the grayscale video frame to an array.
    Vid_Frames.append(gray_frame)


cap.release()
cv2.destroyAllWindows()
```

Retrieve the total number of video frames.

```python
Vid_Frames = np.array(Vid_Frames)

total_frames = Vid_Frames.shape[0]

print("The total number of video frames is %d." % (total_frames))
```

Display a grayscale video frame.

```python
plt.imshow(Vid_Frames[77], cmap='gray')
plt.show()
```
