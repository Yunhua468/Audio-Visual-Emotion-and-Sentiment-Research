# Description: Face Extraction 

This notebook performs the face detection given a video file. The process consists of:
1. Reading a video file
2. Extracting the video frame
3. Converting the video frame to grayscale
4. Detecting the speaker's face in the video frame
5. Extracting the face from the video frame

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Retrieve the name of the video file.

```python
vid_file = 'E:/Video_Speech_Actor_01/Actor_01/01-01-07-01-01-01-01.mp4'
```

Capture the video from the video file.

```python
cap = cv2.VideoCapture(vid_file)
```

Create a face detector by loading a pre-trained classifier for faces.

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

Detect the speaker's face and extract the face from the video frame.

```python
Face_Images = []

while cap.isOpened():
    # Read a video frame.
    ret, frame = cap.read()
    
    # If there is a video frame available for reading.
    if not ret:
        break

    # Convert the video frame from RGB to grayscale.
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect the face in the video frame.
    faces = face_cascade.detectMultiScale(gray_frame, 1.3, 4)
    
    # Extract the face from the video frame.
    for (x, y, w, h) in faces:
        extracted = gray_frame[y:(y + h), x:(x + w)]  
        
    
    # Append the extracted face to an array.
    Face_Images.append(extracted)


cap.release()
cv2.destroyAllWindows()
```

Retrieve the total number of extracted faces.

```python
Face_Images = np.array(Face_Images)

total_extracted = Face_Images.shape[0]

print("The total number of extracted faces is %d." % (total_extracted)) 
```

Display a few faces that were extracted from the video.

```python
for i in range(45, 53):
    plt.imshow(Face_Images[i], cmap='gray')
    plt.show()
```
