# Description: Face Detection 

This notebook performs the face detection given a video file. The process entails reading a video file, extracting the video frame, converting the video frame to grayscale, and detecting the speaker's face.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Retrieve the name of the video file.

```python
vid_file = 'Video_Speech_Actor_01/Actor_01/01-01-05-01-01-01-01.mp4'
```

Capture the video from the video file.

```python
cap = cv2.VideoCapture(vid_file)
```

Create a face detector by loading a pre-trained classifier for faces.

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

Extract the video frames, convert them to grayscale, and detect the speaker's face.

```python
Detected_Frames = []

while cap.isOpened():
    # Read a video frame.
    ret, frame = cap.read()
    
    # If there is a video frame available for reading.
    if not ret:
        break

    # Convert the video frame from RGB to grayscale.
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect the face in the video frame.
    faces = face_cascade.detectMultiScale(gray_frame, 1.3, 4)
    
    # Draw a bounding box within the video frame indicating the detected face.
    for (x, y, w, h) in faces:
        detected = cv2.rectangle(gray_frame, (x, y), (x + w, y + h), (0, 0, 255), 8)   
        
    
    # Append the video frame with the detected face to an array.
    Detected_Frames.append(detected)


cap.release()
cv2.destroyAllWindows()
```

Retrieve the total number of video frames with a detected face.

```python
Detected_Frames = np.array(Detected_Frames)

total_detected = Detected_Frames.shape[0]

print("The total number of video frames with a detected face is %d." % (total_detected)) 
```

Display a video frame with the detected face.

```python
plt.imshow(Detected_Frames[51], cmap='gray')
plt.show()
```
